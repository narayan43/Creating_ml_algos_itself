```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
X= 6*np.random.randn(200,1) - 3
```

```python
y=X**2 + X + 2 + np.random.randn(200,1)*18
```

```python
plt.scatter(X,y,color='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred=model.predict(X_test)
```

```python
print(r2_score(y_pred,y_test))
print(np.sqrt(mean_squared_error(y_pred,y_test)))
```

```python
line=np.linspace(-20,10,200).reshape(200,1)
line_pred=model.predict(line)
```

```python
plt.scatter(X_train, y_train,color='r') 
plt.scatter(X_test, y_test,color='g') 
plt.plot(line, line_pred,color='k') 
plt.show()
```

```python
poly_features=PolynomialFeatures(degree=2)
```

```python
poly_features.fit(X_train)
```

```python
poly_X_train=poly_features.transform(X_train)
poly_X_test=poly_features.transform(X_test)
```

```python
poly_model=LinearRegression()
```

```python
poly_model.fit(poly_X_train,y_train)
```

```python
poly_y_pred=poly_model.predict(poly_X_test)
```

```python
new_x=np.linspace(-20,10,200).reshape(200,1)
new_x_poly= poly_features.transform(new_x)
new_y=poly_model.predict(new_x_poly)
```

```python
plt.scatter(X_train, y_train,color='r') 
plt.scatter(X_test, y_test,color='g') 
plt.plot(new_x, new_y,color='k',linewidth=2) 
plt.show()
```
