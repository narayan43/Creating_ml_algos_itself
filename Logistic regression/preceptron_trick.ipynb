```python
from sklearn.datasets import make_classification
```

```python
X1, Y1 = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1,class_sep=1,random_state=41
)
```

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=X1[:,0],y=X1[:,1],c=Y1,cmap="winter",s=100)

```

```python
def one_zero(d):
    if d>0:
        return 1
    else:
        return 0
```

```python
def perceptron(x,y):
    x =np.insert(x,0,1,axis=1)
    weights=np.ones(x.shape[1])
    lr=0.1
    for i in range(1000):
        j =np.random.randint(0,100)
        y_hat=one_zero(np.dot(x[j],weights))
        weights=weights + lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron(X1,Y1)
```

```python
print(intercept_,coef_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
input_x =np.linspace(-2,2,100)
input_y=m*input_x+b
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X1[:,0],X1[:,1],c=Y1,cmap='winter',s=100)
plt.plot(input_x,input_y,linewidth=3,color='red')
plt.ylim(-3,2)
```

```python



def perceptronTrick(x,y):
    x=np.insert(x ,0,1 ,axis=1)
    lr=.1
    line= np.ones(x.shape[1])
    for i in range(1000):
        randint=np.random.randint(100)
        if np.dot(line,x[randint])>0 and y[randint]==1:
            line =line - lr*(x[randint])
        if np.dot(line,x[randint])<0 and y[randint]==0:
            line =line + lr*(x[randint])
        
    return line[0] ,line[1:]
```

```python
intercept_,coef_=perceptronTrick(X1,Y1)
```

```python
print(intercept_,coef_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
input_x =np.linspace(-2,2,100)
input_y=m*input_x+b
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X1[:,0],X1[:,1],c=Y1,cmap='winter',s=100)
plt.plot(input_x,input_y,linewidth=3,color='red')
plt.ylim(-3,2)
```
